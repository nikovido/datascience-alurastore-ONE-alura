### Importación de datos



```python
import pandas as pd

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

### Preparación de los datos

```python
# Vamos a unir todos los datasets ya que tienen la misma estructura
# Le sumamos una columnna con el valor de su tienda


tienda1["tienda"] = 1
tienda2["tienda"] = 2
tienda3["tienda"] = 3
tienda4["tienda"] = 4

tiendas = pd.concat([tienda1, tienda2, tienda3, tienda4])
tiendas.sample(10)
```

```python
tiendas.info()
```

#1. Análisis de facturación



```python
# Utilizamos el df que contiene todas las tiendas, agrupamos por tienda.
# Ademas vamos a ver los costos de envios, para analizar si hay algun desvio.

ventas_por_tienda = pd.DataFrame(tiendas.groupby("tienda")[["Precio", "Costo de envío"]].sum())

ventas_por_tienda.rename(columns={'Precio': 'ventas', 'Costo de envío': 'envios'}, inplace=True)

ventas_por_tienda.sort_values("ventas", ascending=False, inplace=True)
ventas_por_tienda
```

```python
# Incluyendo envios
ventas_por_tienda.plot.bar(figsize=(10,5))
```

# 2. Ventas por categoría

```python
# Agrupamos por categoria y tienda (vamos a procesar solo las ventas no los envios)
# Le resetiamos el index
ventas_por_categoria= tiendas.groupby(["Categoría del Producto", "tienda"])["Precio"].sum().reset_index()

# Renombramos las columnas para mayor claridad
ventas_por_categoria.rename(
    columns={'Precio': 'ventas', 'Categoría del Producto': 'Categoría'},
    inplace=True
)

# Ordenamos de forma de mayor a menor
ventas_por_categoria.sort_values("ventas", ascending=False, inplace=True)

ventas_por_categoria
```

```python
# Vamos a graficar un poco esto

import seaborn as sns
import matplotlib.pyplot as plt

# Configuramos el estilo de Seaborn
sns.set(style="ticks", context="notebook")

# Setup
g = sns.catplot(
    data=ventas_por_categoria,  #  DataFrame
    x='tienda',                 # Eje X: las tiendas (dentro de cada gráfico)
    y='ventas',                 # Eje Y: las ventas
    col='Categoría',            # Crea la matriz usando esta columna
    kind='bar',                 # Queremos un gráfico de barras
    col_wrap=3,                 # Cuántas columnas de gráficos querés antes de saltar de línea
    sharey=False,               # IMPORTANTE: Que cada gráfico tenga su propia escala Y
    height=4,                   # Aumentamos un poco la altura de cada gráfico para dar más espacio
    aspect=1.2                  # Ajustamos la relación de aspecto para que sean un poco más anchos
)

# Título general de la figura
g.fig.suptitle('Ventas por Categoría y tienda', y=1.03, fontsize=18) # Ajustamos la fuente para que se vea mejor

# Configuramos los títulos de cada subplot
g.set_titles("Categoría: {col_name}")

# Aquí está la CLAVE: Para que todas las etiquetas del eje X sean visibles
# Primero, aseguramos que todas las etiquetas de los ejes se muestren
g.set(xlabel="Tienda", ylabel="Ventas")
g.tick_params(axis='x', labelbottom=True)

plt.tight_layout(rect=[0, 0.03, 1, 0.98]) # Ajusta el layout, dejando espacio para el supertítulo

# Mostramos el gráfico
plt.show()
```

# 3. Calificación promedio de la tienda


```python
# Veamos las calificación promedio de cada tienda
calificacion_promedio = tiendas.groupby(["tienda"])["Calificación"].mean().reset_index()


# Ordenamos de forma de mayor a menor
calificacion_promedio.sort_values("Calificación", ascending=False, inplace=True)

calificacion_promedio
```

```python
calificacion_promedio.plot.bar(x="tienda", y="Calificación", figsize=(10,5))
```

# 4. Productos más y menos vendidos

## Por ingresos

```python
# Primero vamos a analizar los que ingresaron más dinero

# Agrupamos por 'Producto', sumamos el 'Precio' y ordenamos
ingresos_por_producto = tiendas.groupby('Producto')['Precio'].sum().sort_values(ascending=False)

# --- LOS MÁS RENTABLES ---
print("--- TOP 10 MÁS RENTABLES (POR INGRESOS TOTALES) ---")
print(ingresos_por_producto.head(10))

print("\n" + "="*50 + "\n")

# --- LOS MENOS RENTABLES ---
# .tail(10) nos da los 10 productos que MENOS ingresos generaron
print("--- 10 MENOS RENTABLES (POR INGRESOS TOTALES) ---")
print(ingresos_por_producto.tail(10))
```

```python
# Graficamos el TOP 10 de productos por INGRESOS (rentabilidad)
# (usando tu variable 'ingresos_por_producto')

import matplotlib.pyplot as plt
import seaborn as sns

# 1. Asumo que esta variable ya existe de tu paso anterior:
# ingresos_por_producto = tiendas.groupby('Producto')['Precio'].sum().sort_values(ascending=False)

# 2. Creamos el sub-set del TOP 10 a partir de tu variable
top10_ingresos = ingresos_por_producto.head(10)

# 3. Setup de la figura
plt.figure(figsize=(12, 8))

# 4. Graficamos con Seaborn (horizontal)
sns.barplot(
    x=top10_ingresos.values,    # Eje X: Los valores (la suma de 'Precio')
    y=top10_ingresos.index,     # Eje Y: El índice (nombre del 'Producto')
    palette="plasma"            # 'plasma' es otra paleta, para variar
)

# 5. Títulos y etiquetas
plt.title('Top 10 Productos Más Rentables (por Ingresos Totales)', fontsize=16)
plt.xlabel('Ingresos Totales ($)', fontsize=12)
plt.ylabel('Producto', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Mostramos el gráfico
plt.show()
```

## Por unidades vendidas

```python
# Vamos a contar cuántas veces aparece cada producto
# .value_counts() ya los ordena de mayor a menor
productos_por_frecuencia = tiendas['Producto'].value_counts()

# --- LOS MÁS VENDIDOS (TOP 10) ---
print("--- TOP 10 MÁS VENDIDOS (POR UNIDADES) ---")
print(productos_por_frecuencia.head(10))

print("\n" + "="*50 + "\n")

# --- LOS MENOS VENDIDOS (BOTTOM 10) ---
print("--- 10 MENOS VENDIDOS (POR UNIDADES) ---")
print(productos_por_frecuencia.tail(10))
```

```python
# Graficamos el TOP 10 de productos por UNIDADES (frecuencia)

import matplotlib.pyplot as plt
import seaborn as sns

# Creamos el sub-set del TOP 10
top10_unidades = productos_por_frecuencia.head(10)

# Setup de la figura
plt.figure(figsize=(12, 8))

# Graficamos con Seaborn (horizontal)
sns.barplot(
    x=top10_unidades.values,    # Eje X: Los valores (el conteo)
    y=top10_unidades.index,     # Eje Y: El índice (nombre del 'Producto')
    palette="plasma"           # 'viridis' es una buena paleta para conteos
)

# Títulos y etiquetas
plt.title('Top 10 Productos Más Vendidos (por Unidades)', fontsize=16)
plt.xlabel('Número de Ventas (Unidades)', fontsize=12)
plt.ylabel('Producto', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Mostramos el gráfico
plt.show()
```

# 5. Envío promedio por tienda

```python
# Vamos a calcular el costo de envío promedio de cada tienda
envio_promedio = tiendas.groupby(["tienda"])["Costo de envío"].mean().reset_index()

# Ordenamos de forma de mayor a menor para ver qué tiendas tienen los envíos más caros
envio_promedio.sort_values("Costo de envío", ascending=False, inplace=True)

# Vemos los resultados
print("--- Costo de Envío Promedio por Tienda ---")
print(envio_promedio.head())
```

```python
# Graficamos los costos de envío promedio

import matplotlib.pyplot as plt
import seaborn as sns

# 1. Setup de la figura
plt.figure(figsize=(12, 7))

# 2. Graficamos con Seaborn (vertical)
#    Como ya ordenamos el DataFrame, el gráfico respetará ese orden
sns.barplot(
    data=envio_promedio,        # Tu DataFrame ordenado
    x='tienda',                 # Eje X: La tienda
    y='Costo de envío',         # Eje Y: El costo promedio
    palette="coolwarm"          # 'coolwarm' es una paleta que va bien para costos (alto/bajo)
)

# 3. Títulos y etiquetas
plt.title('Costo de Envío Promedio por Tienda', fontsize=16)
plt.xlabel('Tienda', fontsize=12)
plt.ylabel('Costo de Envío Promedio ($)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7) # Grilla horizontal para leer mejor los valores

# Mostramos el gráfico
plt.show()
```
