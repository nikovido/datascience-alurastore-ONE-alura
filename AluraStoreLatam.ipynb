### Importación de datos



```python
import pandas as pd

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

### Preparación de los datos

```python
# Vamos a unir todos los datasets ya que tienen la misma estructura
# Le sumamos una columnna con el valor de su tienda


tienda1["tienda"] = 1
tienda2["tienda"] = 2
tienda3["tienda"] = 3
tienda4["tienda"] = 4

tiendas = pd.concat([tienda1, tienda2, tienda3, tienda4])
tiendas.sample(10)
```

#1. Análisis de facturación



```python
# Utilizamos el df que contiene todas las tiendas, agrupamos por tienda.
# Ademas vamos a ver los costos de envios, para analizar si hay algun desvio.

ventas_por_tienda = pd.DataFrame(tiendas.groupby("tienda")[["Precio", "Costo de envío"]].sum())

ventas_por_tienda.rename(columns={'Precio': 'ventas', 'Costo de envío': 'envios'}, inplace=True)

ventas_por_tienda.sort_values("ventas", ascending=False, inplace=True)
ventas_por_tienda
```

```python
# Incluyendo envios
ventas_por_tienda.plot.bar(figsize=(10,5))
```

# 2. Ventas por categoría

```python
# Agrupamos por categoria y tienda (vamos a procesar solo las ventas no los envios)
# Le resetiamos el index
ventas_por_categoria= tiendas.groupby(["Categoría del Producto", "tienda"])["Precio"].sum().reset_index()

# Renombramos las columnas para mayor claridad
ventas_por_categoria.rename(
    columns={'Precio': 'ventas', 'Categoría del Producto': 'Categoría'},
    inplace=True
)

# Ordenamos de forma de mayor a menor
ventas_por_categoria.sort_values("ventas", ascending=False, inplace=True)

ventas_por_categoria
```

```python
# Vamos a graficar un poco esto

import seaborn as sns
import matplotlib.pyplot as plt

# Configuramos el estilo de Seaborn
sns.set(style="ticks", context="notebook")

# Setup
g = sns.catplot(
    data=ventas_por_categoria,  #  DataFrame
    x='tienda',                 # Eje X: las tiendas (dentro de cada gráfico)
    y='ventas',                 # Eje Y: las ventas
    col='Categoría',            # Crea la matriz usando esta columna
    kind='bar',                 # Queremos un gráfico de barras
    col_wrap=3,                 # Cuántas columnas de gráficos querés antes de saltar de línea
    sharey=False,               # IMPORTANTE: Que cada gráfico tenga su propia escala Y
    height=4,                   # Aumentamos un poco la altura de cada gráfico para dar más espacio
    aspect=1.2                  # Ajustamos la relación de aspecto para que sean un poco más anchos
)

# Título general de la figura
g.fig.suptitle('Ventas por Categoría y tienda', y=1.03, fontsize=18) # Ajustamos la fuente para que se vea mejor

# Configuramos los títulos de cada subplot
g.set_titles("Categoría: {col_name}")

# Aquí está la CLAVE: Para que todas las etiquetas del eje X sean visibles
# Primero, aseguramos que todas las etiquetas de los ejes se muestren
g.set(xlabel="Tienda", ylabel="Ventas")
g.tick_params(axis='x', labelbottom=True)

plt.tight_layout(rect=[0, 0.03, 1, 0.98]) # Ajusta el layout, dejando espacio para el supertítulo

# Mostramos el gráfico
plt.show()
```

# 6. Analisis por fecha

## Las metricas son similares en todas las tiendas, vamos a analizar por fecha para ver las tendencias.

```python
# Convertir la Fecha
tiendas['Fecha de Compra'] = pd.to_datetime(tiendas['Fecha de Compra'], dayfirst=True)

# Agrupar por mes y tienda
tiendas_con_fecha = tiendas.set_index('Fecha de Compra')

# 'M' significa 'Month' (mensual). Agrupamos por tienda Y por mes
# y sumamos el 'Precio'
ventas_por_mes = tiendas_con_fecha.groupby('tienda').resample('M')['Precio'].sum()

# Graficar la película
# Usamos unstack() para convertir el índice 'tienda' en columnas
# Esto crea el gráfico perfecto para comparar
axes = ventas_por_mes.unstack('tienda').plot(
    subplots=True,          # ¡La clave!
    layout=(2, 2),          # Grilla 2x2
    figsize=(15, 10),       # Un tamaño más grande para la grilla
    grid=True,
    sharey=True             # Mantiene el eje Y igual para todas (¡importante!)
)
# -------------------------------------------

# Título general
plt.suptitle('Tendencia de Ventas Mensuales por Tienda', fontsize=16, y=1.02)

# NOTA: Ya no necesitamos .ylabel, .xlabel y .legend()
# porque .plot(subplots=True) los maneja automáticamente
# (y poner un solo título de eje no tendría sentido).

plt.tight_layout(rect=[0, 0.03, 1, 0.98]) # Ajuste para el super-título
plt.show()
```

### Sigamos profundizando, vamos a ver las ventas promedio por año por tienda

```python
# Creamos la nueva columna 'Año'
tiendas['Año'] = tiendas['Fecha de Compra'].dt.year

# Agrupamos por Año y Tienda, calculamos el TICKET PROMEDIO (.mean())
ticket_promedio_anual = tiendas.groupby(['Año', 'tienda'])['Precio'].mean()

# Usamos .unstack() para pivotar la tabla y que sea fácil de leer
# (round(2) es para redondear a 2 decimales, ya que es dinero)
tabla_ticket_promedio = ticket_promedio_anual.unstack('tienda').round(2)

print("--- TICKET PROMEDIO por Año y Tienda ---")
print(tabla_ticket_promedio)
```

```python
# Agrupamos por Año y Tienda, calculamos las VENTAS TOTALES (.sum())
ventas_totales_anual = tiendas.groupby(['Año', 'tienda'])['Precio'].sum()

# Pivotamos la tabla
tabla_ventas_totales = ventas_totales_anual.unstack('tienda').round(0) # Redondeamos a 0 decimales

print("\n--- VENTAS TOTALES por Año y Tienda ---")
print(tabla_ventas_totales)
```

***Insight***: El ticket promedio, asi como las ventas totales de la TIENDA 4 tienden a la baja.

```python
# 'M' significa 'Month' (mensual). Agrupamos por tienda Y por mes
# y calculamos la CALIFICACIÓN PROMEDIO (.mean())
calificacion_por_mes = tiendas_con_fecha.groupby('tienda').resample('M')['Calificación'].mean()

# --- Gráfico de Tendencia de Calificaciones ---
# Usamos la misma lógica de .unstack() y .plot(subplots=True)

axes = calificacion_por_mes.unstack('tienda').plot(
    subplots=True,
    layout=(2, 2),
    figsize=(15, 10),
    grid=True,
    sharey=True,     # ¡Importante! Mantenemos el eje Y (1-5) igual para todas
    style='.-'       # 'style' agrega puntos a la línea, útil para ver los meses
)

# Título general
plt.suptitle('Tendencia de Calificación Promedio Mensual por Tienda', fontsize=16, y=1.02)
plt.tight_layout(rect=[0, 0.03, 1, 0.98])
plt.show()
```

## Analisis por posición

```python
# Vamos a ver cuántos valores únicos de latitud y longitud hay
print(f"Valores únicos de 'lat': {tiendas['lat'].nunique()}")
print(f"Valores únicos de 'lon': {tiendas['lon'].nunique()}")
```

```python
# Vamos a contar, para cada tienda, cuántas zonas (lat) únicas atiende
zonas_por_tienda = tiendas.groupby('tienda')['lat'].nunique().sort_values(ascending=False)

print("--- Zonas Únicas Atendidas por Tienda ---")
print(zonas_por_tienda)
```

```python
# Agrupamos por CADA ZONA (lat y lon) y sumamos el total de ventas
ventas_por_zona = tiendas.groupby(['lat', 'lon'])['Precio'].sum().sort_values(ascending=False)

print("\n--- Ventas Totales por Zona (Top 5) ---")
print(ventas_por_zona.head(5))
```

```python
# Creando la tabla pivote con crosstab
# Filas: Zonas (lat)
# Columnas: Tiendas
# Valores: Suma de Ventas
tabla_zona_tienda = pd.crosstab(
    tiendas['lat'],
    tiendas['tienda'],
    values=tiendas['Precio'],
    aggfunc='sum'
)

# Rellenamos los NaN (zonas donde una tienda no vendió) con 0
tabla_zona_tienda = tabla_zona_tienda.fillna(0)

print("\n--- Tabla Pivote: Ventas de cada Tienda en cada Zona ---")
print(tabla_zona_tienda)
```

```python
# --- 1. Calcular Ingresos y Costos Totales ---
# Agrupamos por tienda y sumamos 'Precio' y 'Costo de envío'
resumen_financiero = tiendas.groupby('tienda')[['Precio', 'Costo de envío']].sum()
resumen_financiero.rename(columns={'Precio': 'Ingresos_Totales', 'Costo de envío': 'Costos_Envio_Totales'}, inplace=True)

# --- 2. Calcular Eficiencia (Ratio de Envío) ---
# ¿Qué porcentaje de la venta se va en el envío? (Más bajo es mejor)
resumen_financiero['Ratio_Envio_%'] = (resumen_financiero['Costos_Envio_Totales'] / resumen_financiero['Ingresos_Totales']) * 100

# --- 3. Calcular Costo Financiero (Cuotas) ---
# ¿Qué tienda financia más? (Más cuotas = más costo para el Sr. Juan)
costo_financiero = tiendas.groupby('tienda')['Cantidad de cuotas'].mean().reset_index()
costo_financiero.rename(columns={'Cantidad de cuotas': 'Cuotas_Promedio'}, inplace=True)

# --- 4. Calificación Promedio (Ya la tenías) ---
# calificacion_promedio = tiendas.groupby(["tienda"])["Calificación"].mean().reset_index()

# --- 5. Unir TODO en una Sola Tabla ---
# (Usamos .set_index('tienda') para poder unirlas)
scorecard = resumen_financiero.merge(
    costo_financiero.set_index('tienda'),
    left_index=True,
    right_index=True
).merge(
    calificacion_promedio.set_index('tienda'),
    left_index=True,
    right_index=True
)

# --- 6. Agregar la Tendencia (Análisis cualitativo) ---
# (Basado en el gráfico ventas-por-fecha.png)
scorecard['Tendencia_Ventas'] = ['Estable', 'Estable', 'Crecimiento', 'Caída']

# --- 7. VER EL SCORECARD FINAL ---
# Ordenamos por la métrica de eficiencia (Ratio_Envio_%)
print("--- SCORECARD DE SALUD DE TIENDAS ---")
print(scorecard.sort_values(by='Ratio_Envio_%', ascending=False))
```

# Informe final: recomendación de cierre de tienda (AluraStore)

---



***1. Introducción y Problema***
El objetivo de este análisis fue determinar cuál de las cuatro tiendas de AluraStore debería ser vendida o cerrada, con el fin de optimizar recursos.

El desafío inicial fue que, en la superficie, todas las tiendas parecían idénticas.

***2. Hallazgo inicial: la "paridad" de las tiendas***
El primer análisis (scorecard de salúd de Ttendas) demostró que las 4 tiendas son clones operativos. Tienen el mismo Ratio de Costo de Envío (todas aprox 5.32%), Calificación Promedio similar (aprox 4.0) y Cuotas Promedio casi idénticas (aprox 2.9).

**Conclusión Parcial**: La eficiencia operativa no es un factor decisivo. La decisión debe tomarse basándose en el rendimiento del mercado.

***3. Hallazgo clave 1: la tendencia de ventas***
El primer punto de quiebre fue el análisis de series de tiempo. Este gráfico mostró el momentum real de cada tienda:

Tienda 3: Muestra una tendencia de crecimiento al final.
Tienda 2: Muestra una caida, luego de una recuperación.
Tiendas 1: Se mantienen estables.
Tienda 4: Es la única tienda que muestra una tendencia de caída sostenida en los últimos meses del período.

**Debilidad Identificada**: La Tienda 4 es la única con una tendencia de ventas negativa, perdiendo valor activamente.

***4. Hallazgo clave 2***: La guerra territorial

El análisis de lat y lon reveló que las 4 tiendas compiten en las mismas 19 zonas. Ventas de cada Tienda en cada Zona nos mostró quién gana esa guerra:

Zonas Clave: El negocio se concentra en zonas de alto valor (ej. 4.60971).

**Debilidad Tienda 4**: En la zona más importante, la Tienda 4 es la que menos vende (419M vs 479M de la líder). En zonas secundarias (ej. 7.89391), la Tienda 4 es aniquilada (vende 7.3M vs 37.8M de la Tienda 1).

**Debilidad Identificada**: La Tienda 4 está perdiendo la batalla en todos los frentes geográficos.

***5. Conclusión y Recomendación***

Se recomienda el cierre o la venta de la Tienda 4.

La decisión no se basa en que sea "menos eficiente" (ya probamos que todas son iguales), sino en dos hechos irrefutables que descubrimos:
- Tendencia: Es el único activo con una tendencia de ventas negativa y sostenida, indicando que sus ventas futuras serán peores.
- Territorio: Es la competidora más débil en las zonas geográficas clave, perdiendo activamente cuota de mercado contra las otras tres tiendas.
